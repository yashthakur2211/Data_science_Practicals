```python
import pandas as pd

# Load the Iris dataset
iris_data = pd.read_csv("iris.csv")
iris_data.head(5)

```

```python
# 1. Summary statistics grouped by 'species'
summary_stats = iris_data.groupby('species')['sepal_length'].describe()

# Add numeric values for each response to the categorical variable
# This is just for reference, so we'll create a dictionary mapping each species to a number
species_map = {
    'Iris-setosa': 1,
    'Iris-versicolor': 2,
    'Iris-virginica': 3
}

# Add a new column 'species_numeric' to the original dataframe
iris_data['species_numeric'] = iris_data['species'].map(species_map)

# Print summary statistics
print("Summary statistics of sepal_length grouped by species:")
print(summary_stats)
print()

# 2. Basic statistical details for each species
# Filter data for each species
setosa_data = iris_data[iris_data['species'] == 'Iris-setosa']['sepal_length']
versicolor_data = iris_data[iris_data['species'] == 'Iris-versicolor']['sepal_length']
virginica_data = iris_data[iris_data['species'] == 'Iris-virginica']['sepal_length']

# Print basic statistical details
print("Basic statistical details for each species:")
print("Iris-setosa:")
print(setosa_data.describe())
print()
print("Iris-versicolor:")
print(versicolor_data.describe())
print()
print("Iris-virginica:")
print(virginica_data.describe())
```
